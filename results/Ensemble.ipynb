```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the clustering data files for each week
clustering_data = pd.read_csv(r"C:\Users\alvar\Documents\Projects\Hydrapig_Analysis\data\Processed\clustering\clustering_data.csv")

# List of variables to plot
variables = [
    'media_ganancia_peso', 'mediana_mililitros', 'mediana_segundos_bebedero', 
    'mediana_segundos_comedero', 'mediana_eventos_bebedero', 'mediana_eventos_comedero'
]

# Plot each variable by cluster evolution over weeks
for var in variables:
    plt.figure(figsize=(10, 6))
    
    for cluster in clustering_data['Cluster'].unique():
        # Filter data for the specific cluster
        cluster_data = clustering_data[clustering_data['Cluster'] == cluster]
        
        # Plot variable over weeks for this cluster
        plt.plot(cluster_data['Week'], cluster_data[var], marker='o', label=f'Cluster {cluster}')
    
    # Graph formatting
    plt.title(f'Evolution of {var} by Cluster')
    plt.xlabel('Week')
    plt.ylabel(var)
    plt.legend(title='Cluster')
    plt.grid(True)
    plt.show()

# Create a bar plot to show the evolution of percentage gain by cluster across weeks
plt.figure(figsize=(12, 8))

# Loop through each cluster and plot its percentage gain across weeks
for cluster in clustering_data['Cluster'].unique():
    cluster_data = clustering_data[clustering_data['Cluster'] == cluster]
    plt.plot(cluster_data['Week'], cluster_data['porcentaje_ganancia_maxima'], marker='o', label=f'Cluster {cluster}')

# Graph formatting
plt.title('Evolution of Percentage Gain by Cluster Across Weeks')
plt.xlabel('Week')
plt.ylabel('Percentage Gain (Max)')
plt.xticks(range(1, 7))  # Assuming we have data for 6 weeks
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```
